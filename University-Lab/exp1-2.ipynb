```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create a social network graph
ng = nx.Graph()
ng.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

# Calculate the degree distribution
degree_sequence = sorted([d for n, d in ng.degree()], reverse=True)
degree_counts = nx.degree_histogram(ng)

# Plot the degree distribution
plt.bar(range(len(degree_counts)), degree_counts)
plt.xlabel('Degree')
plt.ylabel('Count')
plt.title('Degree Distribution (UID - 21BCS2127)')
plt.show() 
```
