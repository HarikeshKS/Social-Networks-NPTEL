```python
import networkx as nx
import matplotlib.pyplot as plt
# for CSV:
G = nx.read_edgelist('DataSets/iris.csv')
print(nx.number_of_nodes(G))
print(nx.number_of_edges(G))
print(nx.is_directed(G))
```

```python
# For Pajek files:
G = nx.read_pajek('DataSets/yeast/Yeast.paj')
print(nx.number_of_nodes(G))
print(nx.number_of_edges(G))
print(nx.is_directed(G))
```

```python
# For graphml
# G = nx.read_graphml('')
print(nx.number_of_nodes(G))
print(nx.number_of_edges(G))
print(nx.is_directed(G))
```

```python
nx.draw_circular(G)
plt.show()
```

```python
# Degree distribution:

nx.degree(G)    # returns a dictionary of the nodes along with its degrees
set(nx.degree(G))    # to print the degrees without redundancy


# Taking out only the degrees not the name of the nodes from the graph for further tasks.
degValList = []
for i,j in nx.degree(G):
    degValList.append(j)
print(degValList)
```

```python
def plot_deg_dist(G):
    all_degrees = degValList    # all the degrees
    unique_degrees = list(set(all_degrees))    # all teh unique degrees
    # aim is the get the count of the unique degrees
    count_of_degrees = []
    for i in unique_degrees:
        x = all_degrees.count(i)
        count_of_degrees.append(x)
    plt.plot(unique_degrees, count_of_degrees, 'go-')
    plt.xlabel('Unique_degrees')
    plt.ylabel('Count_of_degrees')
    plt.title('Degree distribution of some network!')
    plt.show()
# calling the function
plot_deg_dist(G)
```

```python
# Density:
print('The density of above network: ',nx.density(G))

J = nx.complete_graph(100)
print('The density of complete graph: ',nx.density(J))

H = nx.empty_graph()
print('The density of empty graph: ',nx.density(H))
```

```python
# Clustring Coefficient
cln = nx.clustering(G)    # returns a dictionary the has the clustering coefficient values of each node.
for i in cln.items():
    print(i)
nx.average_clustering(G)    # tells about the average clustering, its value is less it means that the network is closely nit.
```

```python
# Diameter:
print('Diameter of the network is : ',nx.diameter(G))
```
